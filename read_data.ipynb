```python
def load_json(filename):
    '''
    Load the given .json files for STAT-628-Module-3
    '''
    res = []
    with open(filename, 'r', encoding='utf-8') as f:
        try:
            while True:
                line = f.readline()
                if line:
                    res.append(json.loads(line))
                else:
                    break
        except:
            f.close()
    res = pd.DataFrame.from_dict(res)
    return res
```
