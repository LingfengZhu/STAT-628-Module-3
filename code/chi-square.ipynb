```
senti = pd.read_csv("senti.csv")
```

```
p_value = []
for i in range(senti.shape[0]):
  
  matrix = np.array(senti.iloc[i,[2,3,4,5]]).reshape(2,2)
  p_value.append(chi2_contingency(matrix)[1])
```

```
senti["p-value"] = p_value
```

```
sum(senti['p-value']>.05)
```

```
senti.to_csv("senti_sig.csv")
```
